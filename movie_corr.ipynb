```python
#importing libraries
import pandas as pd
import numpy
```

```python
movies_df = pd.read_csv("movies.csv")
rating_df = pd.read_csv("ratings.csv")
```

```python
movies_df.head()
```

```python
#removing the year from the title column to make it easier for user to just input the movie name
movies_df['title'] = movies_df['title'].str.replace(r'\s\(\d{4}\)', '', regex=True)
movies_df.head()
```

```python
rating_df.head()
```

```python
#merging the movies_df and ratinf_df on movieId 
df = rating_df.merge(movies_df, on='movieId')
df.head()
```

```python
#pivoting the dataframe to find ratings of titles with userId as index
user_df = df.pivot_table(index = 'userId', columns='title',values='rating')
user_df.head()
```

```python
user_df.info()
user_df.shape
```

```python
#creating a table for titles with their average ratings
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
#counting the no. of votes or ratings a movie has
ratings['no. of ratings'] = df.groupby('title')['rating'].count()
ratings.head()
```

```python
ratings.sort_values('no. of ratings' , ascending=False).head()
```

```python
watching = input("Enter the movie you want recommendation for")
```

```python
#table showing the user rating for the input movie
user_movie=user_df[watching]
user_movie.head()
```

```python
#finding correlation of the input movie with the user_df table
similar_to_user = user_df.corrwith(user_movie)
```

```python
#table in ascending order for highest correlation with input movie
similar_df = pd.DataFrame(similar_to_user,columns=['Correlation'])
similar_df.dropna(inplace=True)
similar_df.sort_values('Correlation',ascending=False).head()
```

```python
similar_df=similar_df.join(ratings['no. of ratings'])
similar_df.head()
```

```python
#filtering the table and removing titles with less than 100 ratings
most_recc=similar_df[similar_df['no. of ratings']>100].sort_values('Correlation',ascending=False)
most_recc.head()
```

```python
result_movie = most_recc.iloc[1:5]
result_movie=result_movie.reset_index()
result_movie.head()
```

```python
#recommendation
result = result_movie['title'].iloc[0:5].tolist()
print("According to the movie you are watching, you should also checout the following movies:")
# result
for i in result:
    print(i)
```
